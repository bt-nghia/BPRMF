```python
movies = pd.read_csv("ml-1m/movies.dat", sep="::",names=["id", "name", "cate"], encoding="ISO-8859-1")
movies
```

```python
s = []
for names in vnm_sbert["req4_data"]:
    x = names.split("|")
    s.append(x)
```

```python
s[0]
```

```python
all_names = []
for i in s:
    mapped_names=[]
    for j in i:
        names, _ = j.split(";")
        mapped_names.append(names.strip())
    all_names.append(mapped_names)
```

```python
all_names[0]
```

```python
all_index = []
for i in all_names:
    p_index = []
    for j in i:
        try:
            p_index.append(movies[movies["name"] == j]["id"].to_numpy()[0])
        except:
            p_index.append(0)
    all_index.append(p_index)
```
